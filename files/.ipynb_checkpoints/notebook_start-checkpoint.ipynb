```python
%matplotlib inline
# import python libraries
import datetime as dt
import gc
import matplotlib.colorbar as cbar
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import matplotlib.lines as mlines


import numpy as np
import os
import pandas as pd
import tkinter as tk
import time

from matplotlib import cm
from nptdms import TdmsFile
from pathlib import Path
from scipy import stats
from scipy.optimize import curve_fit
from scipy.signal import savgol_filter
from tkinter.filedialog import askopenfilename, askdirectory, askopenfilenames

# Define constants
GOLDEN_RATIO = (1 + 5 ** 0.5) / 2

# user defined functions
def get_data_file():
    """Use tkinter to open a file dialog box and allow user selection of """
    """data files. Returns the path of the file."""
    root = tk.Tk()
    file_path = askopenfilename(parent=root,
                                initialdir = r'./', 
                                title = "Select a data file to open ...", 
                                filetypes = (("text files", "*.txt"), ("CSV files", "*.csv"), ("All files", "*.*"))
                                ) # show an "Open" dialog box and return the path to the selected file
    root.quit()
    root.destroy()
    print(''.join([file_path,' selected.']))
    gc.collect()
    return(file_path)

def get_tdms_files():
    """Use tkinter to open a file dialog box and allow user selection of """
    """multiple files. Put files in a list and returns list."""
    root = tk.Tk()
    filez = askopenfilenames(parent=root, 
                             initialdir = r'./', 
#                              filetypes = ('TDMS files', '*.tdms'), 
                             title='Select tdms data files to append to list.'
                            )
    root.quit()
    root.destroy()
    filez = list(filez)
    if len(filez) < 2:
        print(f'The following {len(filez)} file is selected:')
    else:
        print(f'The following {len(filez)} files are selected:')
    for file in filez:
        print(file)
    return(filez) 

def initialize_pressure_transducer(Ilow = 4E-3, Ihigh = 20E-3, Plow = -14.7, Phigh = 45):
    """Initialize the pressure transducer and return values to convert raw data to pressure in psi"""
    """Default values: Ilow = 4E-3, Ihigh = 20E-3, Plow = -14.7, Phigh = 15, R1 = 487.2, R2 = 488.6"""
    # Place ranges into array for easy manipulation for linear model
    Iarray = np.asarray([Ilow, Ihigh])
    Parray = np.asarray([Plow, Phigh])
    # Assume linear fit between the range and calculate the slope and intercept
    LinearModel = stats.linregress(Iarray, Parray)
    slope = LinearModel[0]
    intercept = LinearModel[1]
    print('Pressure transducer initialized:\n\
        I_lo:\t{0:0.3f} Amps\n\
        I_hi:\t{1:0.3f} Amps\n\
        P_lo:\t{2:0.1f} psi\n\
        P_hi:\t{3:0.1f} psi\n\
          '.format(Ilow, Ihigh, Plow, Phigh))
    return slope, intercept

                             
def set_plot_folder():
    """
    Function uses tkinter to promt user to set the plot folder. 
    Printes and returns the selected folder as a path.
    """
    root = tk.Tk()
    path = Path(askdirectory(parent=root,
                             initialdir = r'./',
                             title='Select Plot Folder'))
    root.quit()
    root.destroy()
    print('Plot folder: ', path)
    gc.collect()
    return(path)

def psi_to_Pa(pressure):
    """
    Function returns a presure given in psi to Pa. 
    """
    return pressure * 6894.76

def Pa_to_psi(pressure):
    """
    Function returns a presure given in psi to Pa. 
    """
    return pressure / 6894.76

def K_to_C(temp):
    """Returns a temperature in degrees Celcius given in Kelvin"""
    return temp-272.15

def C_to_K(temp):
    """Returns a temperature in Kelvin given in degrees Celcius"""
    return temp + 272.15

def mmmps_to_mlps(flowrate):
    """Returns a flowrate in ml/s given in m^3/s"""
    return flowrate * 1e6

def mmm_to_ml(volume):
    """Returns a volume in mL given in m^3"""
    return volume * 1e6

def ml_to_mmm(volume):
    """Returns a volume in m^3 given in mL"""
    return volume * 1e-6

def mu_temp_fit(x, a, b, c):
    '''The form of the function that fits viscosity vs temperature data.'''
    return(a * x**b + c)
    
def rho_temp_fit(x, a, b):
    '''The form of the function that fits density vs temperature data.'''
    return(a * x  + b)

def sigma_temp_fit(x, a, b):
    '''The form of the function that fits surface tension vs temperature data.'''
    return(a * x  + b)   
    
def viscosity_temp(temp):
    '''Returns the viscosity (kg/m/s) given a temp (K).'''
    mu_popt_F = np.array([ 5.58586277e+32, -1.38434792e+01,  6.56440863e-03]) 
    mu = mu_temp_fit(temp, *mu_popt_F)
    return mu

def density_temp(temp):
    '''Returns the density (kg/m^3) given a temp (K).'''
    rho_popt_F = np.array([-7.92013082e-01,  1.51830091e+03])
    rho = rho_temp_fit(temp, * rho_popt_F)
    return rho

def surface_tension_temp(temp):
    '''Returns the surface tension (kg/s^2) given a temp (K).'''
    sigma_popt_F = np.array([-6.49547701e-05,  7.34300172e-02])
    sigma = sigma_temp_fit(temp, *sigma_popt_F)
    return sigma

def sum_R_h_lines(temp, L_lines, R_lines):
    '''sums the hydraulic resistance of n lines at temp T in K'''
    R_h_lines = []
    mu = viscosity_temp(temp)
    for line in range(len(L_lines)):
        R_h_lines.append( 8 * mu * L_lines[line] / np.pi / R_lines[line]**4)
    return np.array(R_h_lines).sum()

def R_h_model(temp, N_valve, L_lines, R_lines):
    '''sums and returns the hydraulic resistance of lines and N LFN valves'''
    beta = 6.83e-13    # Units: ?
    mu = viscosity_temp(temp)
    R_h_LFN = N_valve * beta * mu
    R_h_lines = sum_R_h_lines(temp, L_lines, R_lines)
    R_h_total = R_h_LFN + R_h_lines
    return R_h_total

def sum_volume_lines(L_lines, R_lines):
    '''sums the volume of n lines'''
    volume_lines = []
    for line in range(len(L_lines)):
        volume_lines.append(L_lines[line] * np.pi * R_lines[line]**2)
    return np.array(volume_lines).sum()

def doseCalc(x_init, temp, v_dose):
   f        = 1.78;
   k_eff    = 245;
   beta     = 10.76e13;
   a_bore   = 5.86e-5;
   xi       = 0.017;
   mu       = 5.585e32 * (temp ** -1.384e1) + 6.564e-3;
   x_d      = v_dose / a_bore; #checked
   r        = k_eff / (a_bore ** 2 * beta * mu);
   x_star   = x_init + xi - (f / k_eff);
   tau      = -np.log((x_star - x_d) / x_star) / r;
   return   tau

```

# The Accumulator Model


## Measureing spring rate
The spring rate is measured directly using calipers (dial indicator) and a force gauge. The results are are examined below.

|Spring PN | $k_{eff}$ (lb/in)  | $L_E$ (in)  |  $L_C$ (in)  | $k_{eff}$ (N/m)  | $L_E$ (m)  |  $L_C$ (m)  |
|:------|:------:|:------:|:------:|:------::|------:|:------:|
|   S-180  | 1.4 |  1.59  | 0.36 | 245.178 |  40.386e-3  | 9.144e-3 
|   BB-10  | 3.2 | 1.41 | 0.5 | 560.406 | 35.814e-3 | 12.7e-3 |


```python
# The spring rate from manufacturer.
k_ideal = 245.178  # Units: N/m
# We measure the spring rate and account for any differences between the manufacturer and reality.
k_measured = 560.406   # 560.406   #245.178  # Units: N/m
k_loss = k_ideal - k_measured   # Units: N/m
keff = k_ideal - k_loss  # Units: N/m
# The extended and compressed spring lengths are here.
L_spring_extended = 35.81e-3  #35.81e-3  #40.386e-3  # Units: m
L_spring_compressed = 12.7e-3  #12.7e-3  #9.144e-3  # Units: m


L_cap = 3.3e-3    # Units: m
L_piston = 7.6e-3    # Units: m
L_ps = 0    # Units: m
L_p = L_piston - L_ps     # Units: m

# Liquid volume in the wick is calculated below.
diameter_wick = 5.1e-3  # Units: m

df_systems = ['DTB', 'UTB', 'DTB, PMD', 'UTB, PMD']

df_bore_geometry = {'R_bore':[4.32e-3, 4.32e-3, 4.32e-3, 4.32e-3], 'L_bore':[34e-3, 27.5e-3, 34e-3, 27.5e-3]}

df = pd.DataFrame(df_bore_geometry, index=df_systems)
df['L_spring_extended'] = L_spring_extended # Units: m
df['L_spring_compressed'] = L_spring_compressed  # Units: m
df['keff'] = 560.406
df['A_bore'] = np.pi * df['R_bore']**2
df['L_cap'] = 3.3e-3    # Units: m
df['L_piston'] = 7.6e-3    # Units: m
df['L_ps'] = 0    # Units: m
df['L_p'] = df['L_piston'] - df['L_ps']     # Units: m

df['L_spring_squish'] = df[['L_ps','L_spring_compressed']].max(axis=1)  # Units: m
df['L_spring_comp_max'] = df['L_spring_extended'] - df['L_spring_squish']  # Units: m
df['L_spring_preload'] = df['L_bore'] - L_cap - L_p  # Units: m
df['L_spring_comp_min'] = L_spring_extended - df['L_spring_preload'] # Units: m

# Liquid volume of electrode is calculated below.
df['diameter_electrode'] = 2.7e-3  # Units: m
df['length_electrode'] = 19.05e-3  # Units: m
df['porosity_electrode'] = 0.0 # Units: none
df['volume_electrode'] = df['porosity_electrode'] * df['length_electrode'] * np.pi * (df['diameter_electrode'] / 2)**2

# Liquid volume in the wick is calculated below.
df['diameter_wick'] = 5.1e-3  # Units: m
df['length_wick'] = [19.05e-3, 19.05e-3, 5.5e-3, 5.5e-3]  # Units: m
df['porosity_wick'] = 0.5
df['volume_wick'] = df['porosity_wick'] * df['length_wick'] * np.pi * ((df['diameter_wick']/2)**2 - (df['diameter_electrode']/2)**2)

# Liquid volume in porous stack is calculated below.
df['diameter_ps'] = 3.5e-3  # Units: m
df['length_ps'] = 2e-3  # Units: m
df['porosity_ps'] = 0.5
df['volume_ps'] = df['porosity_ps'] * df['length_ps'] * np.pi * (df['diameter_ps']/2)**2

#volume of emitter
df['volume_emitter'] = 20e-9

# The volume of the void in the degasser is here
diameter_sleeve = 7.35e-3  # Units: m
length_sleeve = 15e-3  # Units: m
volume_void_PET = np.pi*((diameter_sleeve/2)**2-(diameter_wick/2)**2) * length_sleeve

df['volume_void'] = [volume_void_PET, volume_void_PET, 264.12e-9, 264.12e-9]  # Units: m^3
  
# Volume of fluid in fluid lines
L_lines_DTB = (2.66e-3, 8.75e-3, 1.64e-3, 7.73e-3, 1e-3, 1e-3, 2.45e-3)
R_lines_DTB = (1.0414e-3/2, 0.5334e-3/2, 1.0414e-3/2, 1.0414e-3/2, 0.53e-3/2, 1e-3/2, 0.3e-3/2)
L_lines_UTB = (5.5e-3, 10.83e-3, 0.49e-3, 0.49e-3, 15.45e-3, 9.33e-3, 1.51e-3, 2.45e-3)
R_lines_UTB = (0.5207e-3, 0.5207e-3, 0.5207e-3, 0.5207e-3, 0.5207e-3, 0.5207e-3, 0.267e-3, 0.3e-3/2)
df['volume_lines'] = [sum_volume_lines(L_lines_DTB, R_lines_DTB), sum_volume_lines(L_lines_UTB, R_lines_UTB), sum_volume_lines(L_lines_DTB, R_lines_DTB), sum_volume_lines(L_lines_UTB, R_lines_UTB)]
df['L_lines'] = [L_lines_DTB, L_lines_UTB, L_lines_DTB, L_lines_UTB]
df['R_lines'] = [R_lines_DTB, R_lines_UTB, R_lines_DTB, R_lines_UTB]

L_lines_DTB_dead = [1.94e-3]
R_line_DTB_dead = [0.27e-3]
volume_DTB_dead = sum_volume_lines(L_lines_DTB_dead,R_line_DTB_dead)

L_lines_UTB_dead = [3.09e-3, 4.59e-3, 1.09e-3]
R_line_UTB_dead = [0.52e-3, 0.52e-3, 0.52e-3]
volume_UTB_dead = sum_volume_lines(L_lines_UTB_dead,R_line_UTB_dead)

df['volume_lines_dead'] = [volume_DTB_dead, volume_UTB_dead, volume_DTB_dead, volume_UTB_dead]

# volume of the LFN valve
df['volume_valve'] = 13e-9 #13ul

# max fill levels
df['volume_max'] = df['A_bore'] * (df['L_bore'] - df['L_cap'] - df['L_piston'] - df['L_spring_squish'])

# Suggested fill volume
df['fill_volume'] = df['volume_void'] + df['volume_lines'] + df['volume_valve']
df['fill_length'] = (df['fill_volume']) / np.pi / df['R_bore']**2

df['fill_rec'] = [.526e-6, .437e-6, .360e-6, .383e-6]

# minimum fill levels
df['volume_min'] = df['volume_wick'] + df['volume_ps'] + df['volume_lines'] + df['volume_emitter'] + df['volume_lines_dead']
df['volume_flood'] = df['volume_void'] + df['volume_lines'] + df['volume_valve'] + df['volume_wick'] + df['volume_ps'] + df['volume_emitter'] + df['volume_lines_dead']



# useable volume
holdup_fraction = 0.5
df['volume_usable'] = df['fill_volume'] - df['volume_lines'] - holdup_fraction * (df['volume_ps'] - df['volume_wick'])

# firing time
fire_rate = 0.82e-12  # Units m^3/s
df['firing_time_s'] = df['volume_usable'] / fire_rate
df['firing_time_m'] = df['firing_time_s'] / 60
df['firing_time_h'] = df['firing_time_m'] / 60
df['firing_time_d'] = df['firing_time_h'] / 24

# weight of propellant
df['propellant_mass'] = df['fill_volume'] * density_temp(C_to_K(23))

# tool length
df['fill_length'] = (df['fill_rec'] / np.pi / df['R_bore']**2)
df['fill_tool_length'] = df['L_bore'] - df['fill_length'] - df['L_piston']


num = 200
df['piston_pos_min'] = 0
df['piston_pos_max'] = df['L_bore'] - df['L_piston'] - df['L_spring_compressed']
df_x_array = pd.DataFrame({'x_array': list(np.linspace(df['piston_pos_min'], df['piston_pos_max'], num=num).transpose())}, index=df_systems)
df = pd.merge(df, df_x_array, how='left', left_index=True, right_index=True)

# account for spring preload
df['x_star']= df['x_array'] + df['L_spring_comp_min']
# account for friction loss
F_loss = 1.78  # Units: N
df['F_spring'] = (df['keff'] * df['x_star']) - F_loss
df['P_spring'] = df['F_spring'] / df['A_bore']
# test accumulator pressure at fill level
df['P_fill'] = (df['keff'] * (df['fill_length'] + df['L_spring_comp_min']) - F_loss) / df['A_bore']

# Temperature ranges 
temp_op_low = C_to_K(15)
temp_op_high = C_to_K(49)
temp = np.linspace(temp_op_low, temp_op_high, num)
df['N_valve'] = 1
# Hyraulic impedance at each temperature
# df['Rh'] = [R_h_model(t, df['N_valve'], df['L_lines'], df['R_lines']) for t in temp]
Rh_UTB = [R_h_model(t, 1, L_lines_UTB, R_lines_UTB) for t in temp]
Rh_DTB = [R_h_model(t, 1, L_lines_DTB, R_lines_DTB) for t in temp]

df_Rh = pd.DataFrame({'Rh': [Rh_DTB, Rh_UTB, Rh_DTB, Rh_UTB]}, index=df_systems)
df = pd.merge(df, df_Rh, how='left', left_index=True, right_index=True)
df['Q'] = df['P_spring'] / df['Rh']
```

```python
# The spring rate from manufacturer.
k_ideal = 245.178  # Units: N/m
# We measure the spring rate and account for any differences between the manufacturer and reality.
k_measured = 560.406   #245.178  # Units: N/m
k_loss = k_ideal - k_measured   # Units: N/m
keff = k_ideal - k_loss  # Units: N/m
# The extended and compressed spring lengths are here.
L_spring_extended = 35.81e-3  #40.386e-3  # Units: m
L_spring_compressed = 12.7e-3  #9.144e-3  # Units: m

# Geometric parameters of the accumulator are here.
D_bore = 4.32e-3 * 2    # Units: m
R_bore = D_bore / 2    # Units: m
A_bore = np.pi * R_bore**2    # Units: m^2
L_bore_DTB = 34.0e-3    # Units: m
L_bore_UTB = 30.7e-3    # Units: m
L_cap = 3.3e-3    # Units: m
L_piston = 7.6e-3    # Units: m
L_ps = 0    # Units: m
L_p = L_piston - L_ps     # Units: m

# Derved geometric variables of the accumulator are here.
L_spring_preload_DTB = L_bore_DTB - L_cap - L_p  # Units: m
L_spring_preload_UTB = L_bore_UTB - L_cap - L_p  # Units: m
L_spring_squish = max(L_ps, L_spring_compressed)  # Units: m
L_spring_comp_min_DTB = L_spring_extended - L_spring_preload_DTB  # Units: m
L_spring_comp_min_UTB = L_spring_extended - L_spring_preload_UTB  # Units: m
L_spring_comp_max = L_spring_extended - L_spring_squish  # Units: m


# The degasser geometry needs to be described here.
# Liquid volume of electrode is calculated below.
diameter_electrode = 2.7e-3  # Units: m
length_electrode = 19.05e-3  # Units: m
porosity_electrode = 0.0 # Units: none
volume_electrode = porosity_electrode * length_electrode * np.pi * (diameter_electrode/2)**2

# Liquid volume in the wick is calculated below.
diameter_wick = 5.1e-3  # Units: m
length_wick_PET = 19.05e-3  # Units: m
length_wick_PMD = 5.5e-3  # Units: m
porosity_wick = 0.5
volume_wick_PET = porosity_wick * length_wick_PET * np.pi * ((diameter_wick/2)**2 - (diameter_electrode/2)**2)
volume_wick_PMD = porosity_wick * length_wick_PMD * np.pi * ((diameter_wick/2)**2 - (diameter_electrode/2)**2)

# Liquid volume in porous stack is calculated below.
diameter_ps = 3.5e-3  # Units: m
length_ps = 2e-3  # Units: m
porosity_ps = 0.5
volume_ps = porosity_ps * length_ps * np.pi * (diameter_ps/2)**2

# The volume of the void in the degasser is here
diameter_sleeve = 7.35e-3  # Units: m
length_sleeve = 15e-3  # Units: m

volume_void_PET = np.pi*((diameter_sleeve/2)**2-(diameter_wick/2)**2) * length_sleeve  # Units: m^3
volume_void_PMD = 264.12e-9

# Volume of fluid in fluid lines
L_lines_DTB = (2.66e-3, 8.75e-3, 1.64e-3, 7.73e-3, 1e-3, 1e-3, 2.45e-3)
R_lines_DTB = (1.0414e-3/2, 0.5334e-3/2, 1.0414e-3/2, 1.0414e-3/2, 0.53e-3/2, 1e-3/2, 0.3e-3/2)
L_lines_UTB = (5.5e-3, 10.83e-3, 0.49e-3, 0.49e-3, 15.45e-3, 9.33e-3, 1.51e-3, 2.45e-3)
R_lines_UTB = (0.5207e-3, 0.5207e-3, 0.5207e-3, 0.5207e-3, 0.5207e-3, 0.5207e-3, 0.267e-3, 0.3e-3/2)

volume_lines_DTB = sum_volume_lines(L_lines_DTB, R_lines_DTB)
volume_lines_UTB = sum_volume_lines(L_lines_UTB, R_lines_UTB)

# Suggested fill volume
fill_volume_DTB = volume_void_PET + volume_lines_DTB
fill_volume_UTB = volume_void_PET + volume_lines_UTB
fill_volume_DTBPMD = volume_void_PMD + volume_lines_DTB
fill_volume_UTBPMD = volume_void_PMD + volume_lines_UTB

fill_length_DTB = (fill_volume_DTB)/np.pi/R_bore**2
fill_length_UTB = (fill_volume_UTB)/np.pi/R_bore**2
fill_length_DTBPMD = (fill_volume_DTBPMD)/np.pi/R_bore**2
fill_length_UTBPMD = (fill_volume_UTBPMD)/np.pi/R_bore**2

# minimum fill levels
volume_min_DTB = volume_wick_PET + volume_ps + volume_lines_DTB
volume_min_UTB = volume_wick_PET + volume_ps + volume_lines_UTB
volume_min_DTBPMD = volume_wick_PMD + volume_ps + volume_lines_DTB
volume_min_UTBPMD = volume_wick_PMD + volume_ps + volume_lines_UTB

# useable volume
volume_usable_DTB = fill_volume_DTB - volume_lines_DTB - volume_ps - volume_wick_PET
volume_usable_UTB = fill_volume_UTB - volume_lines_UTB - volume_ps
volume_usable_DTBPMD = fill_volume_DTBPMD - volume_lines_DTB - volume_ps
volume_usable_UTBPMD = fill_volume_UTBPMD - volume_lines_UTB - volume_ps

fire_rate = 0.82e-12  # Units m^3/s

# firing time
firing_time_DTB = volume_usable_DTB / fire_rate

print(f'Component\t\t Volume (mL)\n\
Electrode\t\t {mmm_to_ml(volume_electrode):0.3f} \n\
Wick (PET) \t\t {mmm_to_ml(volume_wick_PET):0.3f} \n\
Wick (PMD) \t\t {mmm_to_ml(volume_wick_PMD):0.3f} \n\
Porous Stack \t\t {mmm_to_ml(volume_ps):0.3f} \n\
\n\
DGass Void (sans PMD) \t {mmm_to_ml(volume_void_PET):0.3f} \n\
DGass Void (PMD) \t {mmm_to_ml(volume_void_PMD):0.3f} \n\
\n\
Fluid Lines (DTB) \t {mmm_to_ml(volume_lines_DTB):0.3f} \n\
Fluid Lines (UTB) \t {mmm_to_ml(volume_lines_UTB):0.3f} \n\
\n\
Minimum Fill (DTB) \t\t {mmm_to_ml(volume_min_DTB):0.3f} \n\
Minimum Fill (UTB) \t\t {mmm_to_ml(volume_min_UTB):0.3f} \n\
Minimum Fill (DTB, PMD) \t {mmm_to_ml(volume_min_DTBPMD):0.3f} \n\
Minimum Fill (UTB, PMD) \t {mmm_to_ml(volume_min_UTBPMD):0.3f} \n\
\n\
\n\
Suggested Fill (DTB) \t\t {mmm_to_ml(fill_volume_DTB):0.3f} \n\
Suggested Fill (UTB) \t\t {mmm_to_ml(fill_volume_UTB):0.3f} \n\
Suggested Fill (DTB, PMD) \t {mmm_to_ml(fill_volume_DTBPMD):0.3f} \n\
Suggested Fill (UTB, PMD) \t {mmm_to_ml(fill_volume_UTBPMD):0.3f} \n\
\n\
Usable Volume (DTB) \t\t {mmm_to_ml(volume_usable_DTB):0.3f} \n\
Usable Volume (UTB) \t\t {mmm_to_ml(volume_usable_UTB):0.3f} \n\
Usable Volume (DTB, PMD) \t {mmm_to_ml(volume_usable_DTBPMD):0.3f} \n\
Usable Volume (UTB, PMD) \t {mmm_to_ml(volume_usable_UTBPMD):0.3f} \n\
\n\
Firing Time (DTB) \t\t {firing_time_DTB/60/60:0.3f} \n\
Firing Time (UTB) \t\t {mmm_to_ml(volume_usable_UTB):0.3f} \n\
Firing Time (DTB, PMD) \t {mmm_to_ml(volume_usable_DTBPMD):0.3f} \n\
Firing Time (UTB, PMD) \t {mmm_to_ml(volume_usable_UTBPMD):0.3f} \n\
')



```

## Spring Pressure and Maximum Flow Rates

```python
# piston position
num = 200
piston_pos_min = 0
piston_pos_max_DTB = L_bore_DTB-L_piston-L_spring_compressed
x_DTB = np.linspace(piston_pos_min, piston_pos_max_DTB, num=num)
piston_pos_max_UTB = L_bore_UTB-L_piston-L_spring_compressed
x_UTB = np.linspace(piston_pos_min, piston_pos_max_UTB, num=num)

# account for spring preload
x_star_DTB = x_DTB + L_spring_comp_min_DTB
x_star_UTB = x_UTB + L_spring_comp_min_UTB

# account for friction loss
F_loss = 1.78  # Units: N
F_spring_DTB = (keff * x_star_DTB) - F_loss
F_spring_UTB = (keff * x_star_UTB) - F_loss

P_spring_DTB = F_spring_DTB / A_bore
P_spring_UTB = F_spring_UTB / A_bore

N_valve = 1  # number of LFN valves in line

# test accumulator pressure at fill level
P_fill_DTB = ((keff * (fill_length_DTB + L_spring_comp_min_DTB)) - F_loss) / A_bore
P_fill_UTB = ((keff * (fill_length_UTB + L_spring_comp_min_UTB)) - F_loss) / A_bore
P_fill_DTBPMD = ((keff * (fill_length_DTBPMD + L_spring_comp_min_DTB)) - F_loss) / A_bore
P_fill_UTBPMD = ((keff * (fill_length_UTBPMD + L_spring_comp_min_UTB)) - F_loss) / A_bore

print(f'Component\t\t Pressure (psi)\n\
Fill Pressure (DTB) \t\t {Pa_to_psi(P_fill_DTB):0.3f} \n\
Fill Pressure (UTB) \t\t {Pa_to_psi(P_fill_UTB):0.3f} \n\
Fill Pressure (DTB, PMD) \t {Pa_to_psi(P_fill_DTBPMD):0.3f} \n\
Fill Pressure (UTB, PMD) \t {Pa_to_psi(P_fill_UTBPMD):0.3f} \n\
')

# Temperature ranges 
temp_op_low = C_to_K(15)
temp_op_high = C_to_K(49)
temp = np.linspace(temp_op_low, temp_op_high, num)

# Hyraulic impedance at each temperature
Rh_DTB = [R_h_model(t, N_valve, L_lines_DTB, R_lines_DTB) for t in temp]
Rh_UTB = [R_h_model(t, N_valve, L_lines_UTB, R_lines_UTB) for t in temp]


# Maximum fLow rate 
Q_DTB = [P_spring_DTB / R for R in Rh_DTB]
Q_UTB = [P_spring_UTB / R for R in Rh_UTB]
```

# Validating Assumptions
The section below details the validiation of two assumptions:
1. Laminar flow: $\textrm{Re}\equiv\displaystyle\frac{\rho Q D_e}{\mu A_e }<2300$
2. Fully Developed flow: $L_e/L_t<<1$

```python
diameter_exit = 2 * R_lines[-1]
A_exit = np.pi*(diameter_exit/2)**2
# Calculate fluid properties
# density. Unit [kg / m^3]
rho = [density_temp(t) for t in temp]
# viscosity Unit: [kg / m / s]
mu = [viscosity_temp(t) for t in temp]
# surface tension. Unit [kg/s^2]
sigma = [surface_tension_temp(t) for t in temp]

# Calculate Reynolds number.
RE = [keff * x_star / A_bore / Rh[i] / A_exit * diameter_exit * rho[i] / mu[i] for i in range(len(temp))]

# make the plot
fig, ax1 = plt.subplots(1, figsize = [7, 2*GOLDEN_RATIO] , dpi=100)

# plot valve open time as a function of piston position over temperature range.
for i,j in zip(AW, ES):
    ax1.plot(x * 1000, RE[i], c=cm.viridis((j)), zorder=-10)
plt.fill_between([0, fill_length*1000], [150, 150], \
                 alpha=0.1, label=r'Filled Volume $\forall=${0:0.3f} mL'.format(mmm_to_ml(fill_volume)), \
                 color='black', hatch='', linewidth=0.0, zorder=1)
# plt.axhline(2300, color='red')
plt.xlabel(r'Piston Position, $x$ [mm]')    
plt.ylabel(r'$\textrm{Re} = \displaystyle\frac{\rho Q D_e}{\mu A_e } $')
# plt.title(f'Dose = {mmm_to_ml(dose):0.3f} ml, k = {keff:0.0f}, effenciency = {mech_eff:0.2f}')
ax1.set_xlim(0, x.max()*1000)
ax1.set_ylim(0, 150)

# make the color bar
cax, _ = cbar.make_axes(ax1) 
cb2 = cbar.ColorbarBase(cax, cmap=cm.viridis, norm=normal, label='Temperature, $T$ [$^\circ$C]')

# save and show the plot
experiment_id = f'MaxRE_v_Temp'
plot_path = plot_folder.joinpath(''.join(["Plot_", experiment_id, ".", plot_format]))
plt.savefig(plot_path, dpi=600, bbox_inches='tight', transparent=True)
plt.show()
```

# Flow characterization

Calculate and plot the critical nondimensional numbers for the flow.

$\textrm{We}=\dfrac{\rho U^2 L}{\sigma}=\dfrac{\rho Q^2 R_e}{\sigma A_e^2}$

$\textrm{Ca}=\dfrac{\mu U}{\sigma}=\dfrac{\mu Q}{\sigma A_e}$


```python
# Calculate the webber number
WE = [rho[i] * Q[i]**2 / A_exit**2 * diameter_exit/2 / sigma[i] for i in range(len(temp))]

# Calculate the capillary number
CA = [mu[i] * Q[i] / A_exit / sigma[i] for i in range(len(temp))]
```

```python
time_model = np.linspace(0,500,num)

def volumeCalc(time, temp, x_o):
   f        = 1.78;
   k_eff    = 245;
   beta     = 10.76e13;
   a_bore   = 5.86e-5;
   xi       = 0.017;
   mu       = 5.585e32 * (temp ** -1.384e1) + 6.564e-3;
   v_o      = x_o * a_bore; 
   r        = k_eff / (a_bore ** 2 * beta * mu);
   x_star   = x_o + xi - (f / k_eff);
   vol      = v_o - (a_bore * (x_star * np.exp(-r * time) - xi + (f / k_eff)));
   return   vol

# make the plot
fig, ax1 = plt.subplots(1, figsize = [7, 2*GOLDEN_RATIO] , dpi=100)

# plot valve open time as a function of piston position over temperature range.
for i,j in zip(AW, ES):
    ax1.plot(time_model, mmm_to_ml(volumeCalc(time_model, temp[i], fill_length)), c=cm.viridis((j)), zorder=-10)
    
plt.xlabel(r'Time, $t$ [s]')    
plt.ylabel(r'Volume, $\forall$ [mL]')
# plt.title(f'Dose = {mmm_to_ml(dose):0.3f} mL, k = {keff:0.0f} N/m')
ax1.set_xlim(0, time_model.max())
ax1.set_ylim(0, )

# make the color bar
cax, _ = cbar.make_axes(ax1) 
cb2 = cbar.ColorbarBase(cax, cmap=cm.viridis, norm=normal, label='Temperature, $T$ [$^\circ$C]')

# save and show the plot
experiment_id = f'volume_dose_{dose}_keff_{keff}'
plot_path = plot_folder.joinpath(''.join(["Plot_", experiment_id, ".", plot_format]))
plt.savefig(plot_path, dpi=600, bbox_inches='tight', transparent=True)
plt.show()
```
